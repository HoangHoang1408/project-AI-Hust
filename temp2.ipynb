```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import warnings

sns.set()
rcParams['figure.figsize'] = (20,10)
pd.options.display.max_columns = None
warnings.filterwarnings('ignore')
```

```python
move_time = pd.read_csv('./metrics/move_time.csv')

plt.figure(figsize=(20,10))
move_time.groupby('phase')['search_time'].mean().plot(kind='bar')
plt.axhline(0.5, color='red', linestyle='--')
plt.axhline(15, color='green', linestyle='--')
plt.title('Average Search Time by Phase')
```

```python
plt.figure(figsize=(20,10))
move_time.groupby('possible_moves')['search_time'].mean().plot(kind='bar')
plt.axhline(y=0.5, color='r', linestyle='--')
plt.axhline(y=15, color='g', linestyle='--')
plt.title('Average Search Time by Number of possible moves')
```
